```python
import pandas as pd
import json

df = pd.read_parquet(
    "../params/02_prepareeventsds/v011/02_prepareeventsds_dataset.parquet"
)

with open(
    "../params/02_prepareeventsds/v011/02_prepareeventsds_event_catalog.json"
) as f:
    catalog = json.load(f)

nan_codes = {
    code for name, code in catalog.items()
    if name.endswith("_NaN_NaN")
}

count = sum(
    any(ev in nan_codes for ev in evs) if evs is not None else 0
    for evs in df["events"]
)

print("Filas con eventos NaN:", count)
print("Filas con codigos NaN:", nan_codes)
```
